```python
# Desenvolver um programa para solicitar o valor de uma temperatura
# em graus Celsius e convertê-la para graus Fahrenheit.
# A fórmula de conversão é:
# F = C * 9/5 + 32

celsius = float(input("Digite a temperatura em Celsius: "))
fahrenheit = celsius * 9/5 + 32
print("A temperatura em Fahrenheit é: ", fahrenheit, "°F",sep="")
```

# Laboratório 2.6.11



```python
hour = int(input("Hora de início (horas): "))
mins = int(input("Hora de início (minutos): "))
dura = int(input("Duração do evento (minutos): "))

hours_to_min = hour * 60 #converter hora para minutos 
total = hours_to_min + mins + dura #somar tudo

hour_saida = (total // 60) % 24 #aqui eu pego o resto da divisão do total por 24 que é a duração de um dia, só me importa o número inteiro que será as Horas
mins_saida = total % 60 #aqui eu divido o total por 60, só irá me importar o resto que será os Minutos

print(hour_saida,":",mins_saida,sep="")
```
